```python
#importing libraries
import pandas as pd
from pathlib import Path
import numpy as np
import anndata
import time
import matplotlib.pyplot as plt

from abc_atlas_access.abc_atlas_cache.abc_project_cache import AbcProjectCache
from abc_atlas_access.abc_atlas_cache.anndata_utils import get_gene_data
```

```python
#changing path to where data is located locally
download_base = Path('../Documents/data/abc_atlas')
abc_cache = AbcProjectCache.from_cache_dir(download_base)

abc_cache.current_manifest
```

```python
#loading metadata from cells and genes from human dataset
cell = abc_cache.get_metadata_dataframe(
    directory='WMB-10X',
    file_name='cell_metadata',
    dtype={'cell_label': str}
)
cell.set_index('cell_label', inplace=True)
print("Number of cells = ", len(cell))
cell.head(5)
```

```python
# want to see how many unique items associated with each field AND list them out if number is small

def print_column_info(df): 
    for c in df.columns:
        grouped = df[[c]].groupby(c).count()
        members = ''
        if len(grouped) <30:
            members = str(list(grouped.index))
            print("Number of unique %s = %d %s" % (c, len(grouped), members))
        else:
            print("Number of unique %s = %d" %(c, len(grouped)))
```

```python
print_column_info(cell)
```

```python
cluster_details = abc_cache.get_metadata_dataframe(
    directory='WMB-taxonomy',
    file_name='cluster_to_cluster_annotation_membership_pivoted',
    keep_default_na=False
)
cluster_details.set_index('cluster_alias', inplace=True)
cluster_details.head(5)
```

```python
cluster_colors = abc_cache.get_metadata_dataframe(
    directory='WMB-taxonomy',
    file_name='cluster_to_cluster_annotation_membership_color'
)
cluster_colors.set_index('cluster_alias', inplace=True)
cluster_colors.head(5)
```

```python
roi = abc_cache.get_metadata_dataframe(
    directory='WMB-10X',
    file_name='region_of_interest_metadata'
)
roi.set_index('acronym', inplace=True)
roi.rename(columns={'order': 'region_of_interest_order',
                    'color_hex_triplet': 'region_of_interest_color'},
           inplace=True)
roi.head(5)
```

```python
#joining the cluster details, colors, and region of interest onto cell, making it cell_extended
cell_extended = cell.join(cluster_details, on='cluster_alias')
cell_extended = cell_extended.join(cluster_colors, on='cluster_alias')
cell_extended = cell_extended.join(roi[['region_of_interest_order', 'region_of_interest_color']], on='region_of_interest_acronym')
cell_extended.head(5)
```

```python
print_column_info(cell_extended)
```

```python
#function to plot the cells on the map
def plot_umap(xx, yy, cc=None, val=None, fig_width=8, fig_height=8, cmap=None):
    fig, ax = plt.subplots()
    fig.set_size_inches(fig_width, fig_height)
    if cmap is not None:
        plt.scatter(xx, yy, s=0.5, c=val, marker=".",cmap=cmap)
    elif cc is not None:
        plt.scatter (xx, yy, s=0.5, color=cc, marker=".")
    ax.axis('equal')
    ax.set_xlim(-18, 27)
    ax.set_ylim(-18, 27)
    ax.set_xticks([])
    ax.set_yticks([])

    return fig, ax
```

### Creating legend
* find unique pairs of region_of_interest_acronym and region_of_interest_color
* map the pairs to a dictionary
* assign colors and labels in dictionary to patches
* add legend handles into custom legend for image
* optimize using functions where possible

```python
# find unique pairs of roi and roi color
roi_legend = cell_extended[['region_of_interest_acronym','region_of_interest_color']].drop_duplicates().reset_index().drop('cell_label', axis=1)
roi_legend.head()
```

```python
# create dictionary to map the hex codes to - use pd.Series().to_dict()
roi_legend_dict = pd.Series(roi_legend.region_of_interest_acronym.values, index=roi_legend.region_of_interest_color).to_dict()
roi_legend_dict
```

```python
#import mpatches
import matplotlib.patches as mpatches
```

```python
# test for patches - real ones come later

red_patch = mpatches.Patch(color='red', label='The red data')
MOp_patch = mpatches.Patch(color='#179968', label='MOp')
plt.legend(handles=[red_patch,MOp_patch])
```

```python
# smaller dataset to quickly test the plots
cell_subsample = cell_extended.loc[::10]
print(len(cell_subsample))
```

```python
# patch list function to create list of patches for legend from dictionary

# c = color, l = label 
def create_patches (dict):
    patch_list = []
    for c, l in dict.items():
        patch = mpatches.Patch(color=c, label=l)
        patch_list.append(patch)
    return patch_list
```

```python
fig, ax=plot_umap(cell_subsample['x'], cell_subsample['y'], cc=cell_subsample['region_of_interest_color'])
res = ax.set_title("Dissection Region of Interest")

# testing out create_patches
plt.legend(handles=create_patches(roi_legend_dict))
plt.show()
```
